```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.decomposition import PCA
```

```
import matplotlib.patches as mpatches

iris = load_iris() 
X_reduced = PCA(n_components=3).fit_transform(iris.data) 
Y = iris.target

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y, cmap=plt.cm.rainbow_r)

ax.set_title("First three PCA directions") 
ax.set_xlabel("1st eigenvector") 
ax.set_ylabel("2nd eigenvector") 
ax.set_zlabel("3rd eigenvector")

patch = mpatches.Patch(color='red', label='setosa')
patch2 = mpatches.Patch(color='Aqua', label='versicolor')
patch3 = mpatches.Patch(color='purple', label='virginica')
plt.legend(handles=[patch,patch2,patch3])


plt.show()
```
