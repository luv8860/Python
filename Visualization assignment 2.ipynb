# <center> Part 1</center>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

# Building dataset
maxx = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
minn = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

#ploting points
plt.plot(np.arange(12),maxx,'ro')
plt.plot(np.arange(12),minn,'bo')

#defining function to fit the data above in a periodic function
def yearly_temps(times, avg, ampl, time_offset):
    return (avg+ ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

months=np.arange(12)
# Use non-linear least squares to fit a function, f, to data
res_max, cov_max = optimize.curve_fit(yearly_temps, months,maxx, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,minn, [20, 10, 0])

#starting plotting the lines on the fraph
days = np.linspace(0, 12, num=365)

plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(days, yearly_temps(days, *res_min), 'b-')

plt.xlabel('Month')
plt.ylabel('Temperature')

plt.show()
```

```python
import pandas as pd
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
```

Q1.Create a pie chart presenting the male/female proportion

```python
import matplotlib.pyplot as plt
plt.pie(titanic['sex'].value_counts(),labels=['Male','Female'])
plt.legend()
plt.show()
```

Q2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.scatter(titanic["age"][titanic['sex']=="male"],titanic["fare"][titanic['sex']=="male"],label="Male")
plt.scatter(titanic["age"][titanic['sex']=="female"],titanic["fare"][titanic['sex']=="female"],label="Female")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```
